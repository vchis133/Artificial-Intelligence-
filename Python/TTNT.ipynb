```python
# Bài 13:
def bai13():
    # Nhập mảng a
    a = list(map(int, input("Nhập mảng a (các số cách nhau): ").split()))
    # Nhập mảng b
    b = list(map(int, input("Nhập mảng b (các số cách nhau): ").split()))
    result = []
    min_len = min(len(a), len(b))
    for i in range(min_len):
        result.append(a[i] + b[i])
    # Thêm phần tử còn lại
    if len(a) > min_len:
        result.extend(a[min_len:])
    if len(b) > min_len:
        result.extend(b[min_len:])
    print("Mảng kết quả:", result)
bai13()
```

```python
# Bài 14:
import random
def bai14():
    m = int(input("Nhập số dòng m: "))
    n = int(input("Nhập số cột n: "))
    # a. Tạo mảng a chứa các số nguyên ngẫu nhiên (giả sử từ 0 đến 100)
    a = [[random.randint(0, 100) for _ in range(n)] for _ in range(m)]
    print("Mảng a:")
    for row in a:
        print(row)

    # b. Xuất các phần tử thuộc dòng k
    k = int(input("Nhập dòng k (0-based): "))
    if 0 <= k < m:
        print(f"Dòng {k}:", a[k])
    else:
        print("Dòng không hợp lệ")

    # c. Xuất các phần tử thuộc cột k
    k = int(input("Nhập cột k (0-based): "))
    if 0 <= k < n:
        col = [a[i][k] for i in range(m)]
        print(f"Cột {k}:", col)
    else:
        print("Cột không hợp lệ")

    # d. Tìm dòng có tổng lớn nhất
    sums = [sum(row) for row in a]
    max_sum = max(sums)
    max_row = sums.index(max_sum)
    print(f"Dòng có tổng lớn nhất: dòng {max_row} với tổng {max_sum}")
    # e. Tìm cột có tích nhỏ nhất (tích các phần tử)
    products = []
    for j in range(n):
        prod = 1
        for i in range(m):
            prod *= a[i][j]
        products.append(prod)
    min_prod = min(products)
    min_col = products.index(min_prod)
    print(f"Cột có tích nhỏ nhất: cột {min_col} với tích {min_prod}")

    # f. Xuất các phần tử thuộc dòng chẵn và cột lẻ (chỉ số chẵn/lẻ)
    print("Các phần tử thuộc dòng chẵn và cột lẻ:")
    for i in range(m):
        if i % 2 == 0:
            for j in range(n):
                if j % 2 == 1:  # cột lẻ
                    print(f"a[{i}][{j}] = {a[i][j]}")

    # g. Tính trung bình cộng các phần tử chẵn thuộc dòng lẻ
    even_sum = 0
    count = 0
    for i in range(m):
        if i % 2 == 1:
            for j in range(n):
                if a[i][j] % 2 == 0:
                    even_sum += a[i][j]
                    count += 1
    if count > 0:
        avg = even_sum / count
        print(f"Trung bình cộng các phần tử chẵn thuộc dòng lẻ: {avg}")
    else:
        print("Không có phần tử chẵn trên dòng lẻ")

    # h. Tính trung bình cộng các phần tử thuộc biên
    border_sum = 0
    border_count = 0
    for i in range(m):
        for j in range(n):
            if i == 0 or i == m-1 or j == 0 or j == n-1:
                border_sum += a[i][j]
                border_count += 1
    if border_count > 0:
        avg_border = border_sum / border_count
        print(f"Trung bình cộng các phần tử biên: {avg_border}")
    else:
        print("Không có phần tử biên")

    # i. Tính trung bình tích các phần tử không thuộc biên
    # Tích các phần tử không thuộc biên
    product = 1
    non_border_count = 0
    for i in range(m):
        for j in range(n):
            if not (i == 0 or i == m-1 or j == 0 or j == n-1):
                product *= a[i][j]
                non_border_count += 1
    if non_border_count > 0:
        # Trung bình tích (mean of product?) Thực chất là tích các phần tử, nhưng đề nói "trung bình tích" có thể là tích các phần tử? Đọc lại: "Tính trung bình tích các phần tử không thuộc biên." Có thể hiểu là tích của tất cả các phần tử đó? Nhưng "trung bình tích" thì không rõ. Có thể là trung bình cộng của các tích? Thông thường, "tính trung bình tích" là tính tích các phần tử? Tôi sẽ tính tích.
        print(f"Tích các phần tử không thuộc biên: {product}")
    else:
        print("Không có phần tử không thuộc biên")

bai14()
```

```python

# Bài 15:
def bai15():
    # a. Xây dựng cấu trúc SINHVIEN (dùng class hoặc namedtuple)
    class SinhVien:
        def __init__(self, ma, ten, nam_sinh, diem_tb):
            self.ma = ma
            self.ten = ten
            self.nam_sinh = nam_sinh
            self.diem_tb = diem_tb

    # b. Cho một mảng có n sinh viên (tạo mẫu)
    n = int(input("Nhập số lượng sinh viên: "))
    ds = []
    for i in range(n):
        print(f"Nhập thông tin sinh viên thứ {i+1}:")
        ma = input("Mã sinh viên (10 ký tự): ")
        ten = input("Tên (tối đa 20 ký tự): ")
        nam_sinh = int(input("Năm sinh: "))
        diem_tb = float(input("Điểm trung bình: "))
        ds.append(SinhVien(ma, ten, nam_sinh, diem_tb))

    # c. Hàm cho biết có bao nhiêu sinh viên đủ điều kiện lên lớp (điểm >=5)
    def so_sv_len_lop(ds):
        count = 0
        for sv in ds:
            if sv.diem_tb >= 5:
                count += 1
        return count
    print(f"Số sinh viên đủ điều kiện lên lớp: {so_sv_len_lop(ds)}")

    # d. Xuất các sinh viên đủ 20 tuổi
    import datetime
    current_year = datetime.datetime.now().year
    tuoi_20 = current_year - 20
    print("Các sinh viên đủ 20 tuổi (sinh năm <= {tuoi_20}):")
    for sv in ds:
        if sv.nam_sinh <= tuoi_20:
            print(f"Mã: {sv.ma}, Tên: {sv.ten}, Năm sinh: {sv.nam_sinh}, Điểm: {sv.diem_tb}")

    # e. Đếm số sinh viên học hệ đại học (mã chứa 'DH' ở vị trí 2,3)
    def dem_he_dh(ds):
        count = 0
        for sv in ds:
            if len(sv.ma) >= 4 and sv.ma[2:4] == "DH":
                count += 1
        return count
    print(f"Số sinh viên hệ đại học: {dem_he_dh(ds)}")

    # f. Đếm số sinh viên có tên "Lan"
    def dem_ten_lan(ds):
        count = 0
        for sv in ds:
            if sv.ten.strip() == "Lan":
                count += 1
        return count
    print(f"Số sinh viên tên Lan: {dem_ten_lan(ds)}")

    # g. Đếm số sinh viên có họ "Phan"
    def dem_ho_phan(ds):
        count = 0
        for sv in ds:

            ho = sv.ten.strip().split()[0] if sv.ten.strip() else ""
            if ho == "Phan":
                count += 1
        return count
    print(f"Số sinh viên họ Phan: {dem_ho_phan(ds)}")

bai15()
```
