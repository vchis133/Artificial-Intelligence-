```python
import math

N = 5

def create_board():
    return [["_" for _ in range(N)] for _ in range(N)]

def evaluate(board):
    score = 0
    for row in board:
        score += row.count("X") - row.count("O")
    for col in range(N):
        col_vals = [board[row][col] for row in range(N)]
        score += col_vals.count("X") - col_vals.count("O")
    return score

def alphabeta(board, depth, alpha, beta, is_max):
    if depth == 0:
        return evaluate(board)

    if is_max:
        best = -math.inf
        for i in range(N):
            for j in range(N):
                if board[i][j] == "_":
                    board[i][j] = "X"
                    val = alphabeta(board, depth - 1, alpha, beta, False)
                    board[i][j] = "_"
                    best = max(best, val)
                    alpha = max(alpha, best)
                    if beta <= alpha:
                        return best
        return best
    else:
        best = math.inf
        for i in range(N):
            for j in range(N):
                if board[i][j] == "_":
                    board[i][j] = "O"
                    val = alphabeta(board, depth - 1, alpha, beta, True)
                    board[i][j] = "_"
                    best = min(best, val)
                    beta = min(beta, best)
                    if beta <= alpha:
                        return best
        return best
board = create_board()
board[0][0] = "X"
board[1][1] = "O"

print("=== ALPHA-BETA NxN (5x5) ===")
result = alphabeta(board, 2, -math.inf, math.inf, True)
print(f"Giá trị đánh giá: {result}")
```

```python
import chess
import math

board = chess.Board()

def evaluate(board):
    piece_values = {
        chess.PAWN: 1,
        chess.KNIGHT: 3,
        chess.BISHOP: 3,
        chess.ROOK: 5,
        chess.QUEEN: 9
    }

    score = 0
    for piece_type in piece_values:
        score += len(board.pieces(piece_type, chess.WHITE)) * piece_values[piece_type]
        score -= len(board.pieces(piece_type, chess.BLACK)) * piece_values[piece_type]

    return score

def alphabeta(board, depth, alpha, beta, is_max):
    if depth == 0 or board.is_game_over():
        return evaluate(board)

    if is_max:
        best = -math.inf
        for move in board.legal_moves:
            board.push(move)
            best = max(best, alphabeta(board, depth-1, alpha, beta, False))
            board.pop()
            alpha = max(alpha, best)
            if beta <= alpha:
                break
        return best
    else:
        best = math.inf
        for move in board.legal_moves:
            board.push(move)
            best = min(best, alphabeta(board, depth-1, alpha, beta, True))
            board.pop()
            beta = min(beta, best)
            if beta <= alpha:
                break
        return best

print("=== CHESS AI ===")
print(board)

print("Đánh giá:", alphabeta(board, 2, -math.inf, math.inf, True))
```
