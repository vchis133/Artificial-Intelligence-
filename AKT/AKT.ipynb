```python
import heapq
import copy

N = 5  # thay đổi tùy ý (5x5)

def create_goal(n):
    goal = list(range(1, n*n)) + [0]
    return [goal[i*n:(i+1)*n] for i in range(n)]

GOAL = create_goal(N)

def find_zero(state):
    for i in range(N):
        for j in range(N):
            if state[i][j] == 0:
                return i, j

def heuristic(state):
    dist = 0
    for i in range(N):
        for j in range(N):
            val = state[i][j]
            if val != 0:
                x = (val-1)//N
                y = (val-1)%N
                dist += abs(x-i) + abs(y-j)
    return dist

def get_neighbors(state):
    x, y = find_zero(state)
    moves = [(-1,0),(1,0),(0,-1),(0,1)]
    neighbors = []

    for dx, dy in moves:
        nx, ny = x+dx, y+dy
        if 0 <= nx < N and 0 <= ny < N:
            new_state = copy.deepcopy(state)
            new_state[x][y], new_state[nx][ny] = new_state[nx][ny], new_state[x][y]
            neighbors.append(new_state)
    return neighbors

def astar(start):
    pq = []
    heapq.heappush(pq, (0, start, 0))
    visited = set()

    while pq:
        f, state, g = heapq.heappop(pq)

        if str(state) in visited:
            continue
        visited.add(str(state))

        if state == GOAL:
            print("Solved in steps:", g)
            return

        for neighbor in get_neighbors(state):
            if str(neighbor) not in visited:
                h = heuristic(neighbor)
                heapq.heappush(pq, (g + 1 + h, neighbor, g + 1))
```
