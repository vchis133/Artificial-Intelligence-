```python
import heapq

# Ma trận khoảng cách
dist = [
    [0, 10, 15, 20],
    [10, 0, 35, 25],
    [15, 35, 0, 30],
    [20, 25, 30, 0]
]

N = len(dist)

def heuristic(node, visited):
    return 0  # đơn giản (có thể cải tiến)

def tsp_astar():
    pq = []
    heapq.heappush(pq, (0, 0, [0], 0))  # f, node, path, cost

    while pq:
        f, node, path, cost = heapq.heappop(pq)

        if len(path) == N:
            return path + [0], cost + dist[node][0]

        for i in range(N):
            if i not in path:
                new_cost = cost + dist[node][i]
                h = heuristic(i, path)
                heapq.heappush(pq, (new_cost + h, i, path + [i], new_cost))

# Test
path, cost = tsp_astar()
print("Đường đi:", path)
print("Chi phí:", cost)
```
